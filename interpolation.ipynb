```python
from scipy.interpolate import interp1d
from numpy import *
import matplotlib.pyplot as plt
```

### Problem 1 - Interpolation using scipy.interpolatete

```python
list = loadtxt('scattering.data', float)
```

```python
x, y = list[:, 0], list[:, 1]
```

```python
time = linspace(0, 180, 180)

f = interp1d(x, y)
f2 = interp1d(x, y, kind = 'cubic')

listXs_linear = f(time)
listXs_cubic = f2(time)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (12, 4))

axs[0].plot(listXs_linear)
axs[0].set_title('linear interpolation')
axs[0].set_xlabel('Energy (MeV)')
axs[0].set_ylabel('Cross section')
axs[0].plot(x, y, '.')

axs[1].plot(listXs_cubic)
axs[1].set_title('cubic interpolation')
axs[1].set_xlabel('Energy (MeV)')
axs[1].plot(x, y, '.')

plt.show()
```

### Problem 2 - linear interpolation

```python
def func(i, xi):
    num = y[i-1] * (x[i] - xi) + y[i] * (xi - x[i-1])
    den = x[i] - x[i - 1]
    
    return num/den
```

```python
x2, y2 = list[:, 0], list[:, 1]
```

```python
intervals = []
func_values = []

for i in range(1, len(x)):
    one_interval = linspace(x[i-1], x[i], 10)
    
    for j in range(len(one_interval)):
        
        f = func(i, one_interval[j])
        intervals.append(one_interval[j])
        func_values.append(f)
        
plt.plot(intervals, func_values)
plt.title('linear interpolation')
plt.xlabel('Energy (MeV)')
plt.ylabel('Cross section')
plt.plot(x2, y2, '.')

plt.show()
```
